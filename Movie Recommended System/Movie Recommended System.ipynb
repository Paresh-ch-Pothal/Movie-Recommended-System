```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
        
```

```python
# to make the id integer we do the following
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
# for crew
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break
    return l
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"]=movies["genres"].apply(lambda x : [i.replace(" ","") for i in x])
movies["keywords"]=movies["overview"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x : [i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
new_df=movies[["movie_id","title","tags"]]
```

```python
# converting the tags list into string
new_df["tags"]=new_df["tags"].apply(lambda x: " ".join(x))
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

# Here basically we are doing that ["loved","Loving","Love'] all becomes ==> ["love","love","love"]
        
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
new_df["tags"][0]
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# Text Vectorization : it is mainly used to find the similarity between the text sentences
# here The idea is Concatenating of all the tags and i mainly find out the 5000 most common
# words .
# After that i mainly search in each tag of a sngle row how many times these 5000 most
# common words are present .. Similarly i calculate for every tag and i got a 5000x5000
# dataset and all rows in this dataset is acts as a vector
# here we are not use any stop words ex: are,is,of,the,it,then,from,him,to etc .... 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
# To see the 5000 features
cv.get_feature_names_out()
```

```python
# Here we have to calculate the distance between the vectors so that the lowest distance 
# vectors is more similar to other .
# Here instead of calculating the euclidean distance we mainly calculate the angle
# between the two vectors . ---> Less angle ===> More Similar
# More angle ===> Less Similar
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1]) # Here iam saying
# to sort the list according to the 1st index because this enumerate function creates a 
# list which stores index at 0th index and value at the 1st index and we want to sort the
# list according to the value not according to the index
```

```python
def recommend(movie):
    # To find out the index of a given movie
    movie_index=new_df[new_df["title"] == movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman")
```
